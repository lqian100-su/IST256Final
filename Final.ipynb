```python
# time zone finder, Justin Wang
#import allllll the goodies!
import os
from collections import defaultdict
from datetime import datetime
import urllib.request
import urllib.parse
from zipfile  import ZipFile
import pytz 
#identify file
url = 'http://download.geonames.org/export/dump/'
basename = 'cities15000' 
filename = basename + '.zip'

# get the file using urllib
if not os.path.exists(filename):
    urllib.request.urlretrieve(urllib.request.urljoin(url, filename), filename)

# parse the file to do timezone finding
timezones = defaultdict(set)
with ZipFile(filename) as zf, zf.open(basename + '.txt') as file:
    for line in file:
        values = line.split(b'\t')
        if values: 
            name, asciiname, alternatenames = values[1:4]
            timezone = values[-2].decode('utf-8').strip()
            if timezone:
                for city in [name, asciiname] + alternatenames.split(b','):
                    city = city.decode('utf-8').strip()
                    if city:
                        timezones[city].add(timezone)



# input and output
formats = '%Y-%m-%d %H:%M:%S %Z%z'
city = input("Enter city name: ")
for tzname in timezones[city]:
    now = datetime.now(pytz.timezone(tzname))
    print("")
    print("%s is in %s timezone" % (city, tzname))
    print("Current time in %s is %s" % (city, now.strftime(formats)))


#resturant finder Justin Wang
import json
import requests
from pprint import pprint
city = input("Enter city name: ")
url = "https://developers.zomato.com/api/v2.1/cities?q=" + city
header = {"User-agent": "curl/7.43.0", "Accept": "application/json", "user_key": "18be42a2ada2f9e0fa409457579a66a0"}

response = requests.get(url, headers=header)
data = response.json()
city_id = str(data['location_suggestions'][0]['id'])

url2 = ("https://developers.zomato.com/api/v2.1/location_details?entity_id= %s&entity_type=city" % (city_id))
header = {"User-agent": "curl/7.43.0", "Accept": "application/json", "user_key": "18be42a2ada2f9e0fa409457579a66a0"}
response2 = requests.get(url2, headers=header)
best = response2.json()
pprint(best)
```
